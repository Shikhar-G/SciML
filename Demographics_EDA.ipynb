```python
import pandas as pd

# Loading the dataset
df_demographics = pd.read_csv('/content/machine_learning_TJ_v7_demographics (1).csv')

# Displaying the first few rows of the dataframe
df_demographics.head()
```

```python
# Checking for missing values
df_missing = df_demographics.isnull().sum()

# Generating descriptive statistics for numerical features
df_describe = df_demographics.describe()

# Outputting the results
print('Missing Values:\
', df_missing)
print('\
Descriptive Statistics:\
', df_describe)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure
plt.figure(figsize=(20, 10), facecolor='white')

# Plotting distributions of categorical variables
plt.subplot(2, 3, 1)
sns.countplot(x='race_WHITE', data=df_demographics)
plt.title('Distribution of White Race')

plt.subplot(2, 3, 2)
sns.countplot(x='gender_M', data=df_demographics)
plt.title('Distribution of Male Gender')

plt.subplot(2, 3, 3)
sns.countplot(x='financial_class_MEDICARE', data=df_demographics)
plt.title('Distribution of Medicare Financial Class')

plt.subplot(2, 3, 4)
sns.countplot(x='ethnicity_NOT HISPANIC OR LATINO', data=df_demographics)
plt.title('Distribution of Non-Hispanic or Latino Ethnicity')

plt.subplot(2, 3, 5)
sns.countplot(x='admit_quarter', data=df_demographics)
plt.title('Admission Quarter Distribution')

plt.tight_layout()
plt.show()
```

```python
# Plotting the distribution of age
plt.figure(figsize=(10, 6), facecolor='white')
sns.histplot(df_demographics['age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution in the Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
